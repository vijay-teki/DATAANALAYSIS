```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz,plot_tree
from IPython.display import Image
```

```python
df=pd.read_csv('/content/drive/MyDrive/dataset/demoDT (1).csv',sep=",")
print(df)
```

```python
feature_cols = ['Literacy',' Cleanliness',' Crime_Rate']
feature=df[feature_cols]
Target = df.Good
```

```python
DT=DecisionTreeClassifier()
DT=DT.fit(feature,Target)
```

```python
l=int(input('Enter Literacy'))
cl=int(input("enter Cleanliness"))
cr=int(input("enter crime rate"))
pred=DT.predict([[l,cl,cr]])
if pred==1:
  print("good state")
else:
  print("bad state")
```

```python
plot_tree(DT)
```
